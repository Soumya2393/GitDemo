```python
# Q6) In customer table, create a new variable called â€œcustomer value segmentâ€ using customer value as follows.
#     - High Value Segment - > 25000
#     - Medium Value Segment â€“ Between 10000 and 25000
#     - Low Value Segment â€“ less than or equal to 10000
```

```python
Customers['Customer_Value_Segment'] = ""
np.where(Customers.Customer_Value >= 25000)
Customers['Customer_Value_Segment'] = "High Value Segment"
np.where((Customers.Customer_Value > 10000)&(Customers.Customer_Value < 25000))
Customers['Customer_Value_Segment'] = "Medium Value Segment"
np.where(Customers.Customer_Value <= 10000)
Customers['Customer_Value_Segment'] = "Low Value Segment"
```
